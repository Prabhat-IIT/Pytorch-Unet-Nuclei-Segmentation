```python
target = io.imread('./he.png')
```

```python
m = MacenkoNormalizer()
r = ReinhardNormalizer()
v = VahadaneNormalizer()
m.fit(target)
r.fit(target)
v.fit(target)
```

```python
n = '15'
source = io.imread('./train-img128x128/'+n+'_0.png')
mask = io.imread('./train-mask128x128/'+n+'_0.png')
```

```python
out1 = m.transform(source)
out2 = r.transform(source)
out3 = v.transform(source)
obw1 = m.hematoxylin(source)
obw3 = v.hematoxylin(source)
```

```python
normalized = build_stack((source,out1,out2,out3,mask))
hemo = build_stack((obw1,obw3))
```
